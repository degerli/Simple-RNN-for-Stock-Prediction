```python
''' We will create a recurrent neural network (RNN) using stock data as input'''
import os
import numpy as np, pandas as pd, quandl, tensorflow as tf, Get_and_Clean_Stock_Data as clean, RNN_Training as RNN_train
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3' 
```

```python
''' To get data directly from quandl, then clean the data appropriately. '''
file_name = 'top_stocks'
csv_file = file_name + '.csv'

'''
example of get_clean_to_csv:

['F']: Stock to get from Quandl
8, 12: Columns of the dataframe taken from Quandl (8-12 is adjusted OHLC, volume)
.95: Split percentage of train set and test set
top_stocks: name used to label the created csv files for test and train
'''

#train = clean.get_clean_to_csv (['F'], 8, 12, .95, file_name)
train = pd.read_csv('CleanedData/' + file_name + '_train.csv')
train.set_index("Date", inplace=True)

test = pd.read_csv('CleanedData/' + file_name + '_test.csv')
test.set_index("Date", inplace=True)

train.head()
```

```python
n_steps = 10
n_neurons = 50
n_layers = 3
learning_rate = 0.001
n_iterations = 1500
batch_size = 50
train_keep_prob = 0.5

outputs, states, saver, X, y = RNN_train.create_RNN_layers (train, n_steps, train.columns, n_neurons, n_layers, learning_rate, n_iterations, batch_size, train_keep_prob, file_name)
# X, y are the placeholders
```

```python
n_future_iter = 1
plot_column = train.columns[-2]
sequence = RNN_train.create_predictions_for_entire_test_set (test, plot_column, n_steps, outputs, n_future_iter, file_name, saver, X, y)
#sequence
```

```python
import matplotlib.pyplot as plt, matplotlib.patches as mpatches
% matplotlib notebook
test = pd.read_csv('CleanedData/' + file_name + '_test.csv')
test.set_index("Date", inplace=True)
actual_test_data = test[plot_column][:].values

plt.title("Checking if Test Stocks Match Actual Stocks", fontsize=12)
plt.plot(actual_test_data, 'g', label='Actual Data')
plt.plot(sequence, 'r--')
red_patch = mpatches.Patch(color='red', label='Predicted Data')
green_patch = mpatches.Patch(color='green', label='Actual Data')

plt.legend(handles=[red_patch, green_patch])
plt.axhline(y=0,)
plt.show()
```
